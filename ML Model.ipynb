```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Car Data.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset["year"].unique
```

```python
dataset = dataset[dataset["year"].str.isnumeric()]
```

```python
dataset["year"] = dataset["year"].astype(int)
```

```python
dataset = dataset[dataset["Price"] != "Ask For Price"]
```

```python
dataset["Price"] = dataset["Price"].str.replace(",","").astype(int)
```

```python
dataset["kms_driven"] = dataset["kms_driven"].str.split(" ").str.get(0).str.replace(",","")
```

```python
dataset = dataset[dataset["kms_driven"].str.isnumeric()]
```

```python
dataset["kms_driven"] = dataset["kms_driven"].astype(int)
```

```python
dataset.dropna(inplace = True)
```

```python
dataset["name"] = dataset["name"].str.split(" ").str.slice(0,3).str.join(" ")
```

```python
dataset
```

```python
dataset.describe()
```

```python
dataset = dataset[dataset["Price"] < 4e6]
```

```python
dataset.info()
```

```python
dataset = dataset.reset_index(drop = True)
```

```python
dataset.info()
```

```python
dataset.to_csv("Clean Car Dataset.csv")
```

```python
x = dataset.drop(columns = "Price")
y = dataset["Price"]
```

```python
from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[["name", "company", "fuel_type"]])  
```

```python
ohec_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ["name", "company", "fuel_type"]), remainder = "passthrough")
```

```python
scores = []
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = i)
    lr = LinearRegression()
    carpipeline = make_pipeline(ohec_trans, lr)
    carpipeline.fit(x_train, y_train)
    y_pred = carpipeline.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
    
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = np.argmax(scores))
lr = LinearRegression()
carpipeline = make_pipeline(ohec_trans, lr)
carpipeline.fit(x_train, y_train)
y_pred = carpipeline.predict(x_test)
r2_score(y_test, y_pred)
```
